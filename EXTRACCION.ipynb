```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("Detalle precios y productos fabricados 2022.xlsx")
df
```

```python
#Se aplica filtro a la columna nombre_vendador para encontrar a todas las LETICIA RAMIREZ HERNANDEZ
filtro1=df[df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ"]
filtro1
```

```python
filtro1.to_csv("filtro1.csv")
from google.colab import files
files.download("filtro1.csv")
```

```python
#SE APLICA FILTRO PARA PRESENTAR LAS FILAS 8, 24 Y 100 CON TODAS LAS COLUMNAS
filtro2=df.iloc[ [8, 24, 100] , : ]  #Filas no consecutivas
filtro2
```

```python
filtro2.to_csv("filtro2.csv")
from google.colab import files
files.download("filtro2.csv")
```

```python
#SE APLICA FILTRO A TODAS LAS FILAS SOLO CON LAS COLUMNAS 1,4 Y 6
filtro3=df.iloc[ : , [1,4,6]]  #Columnas no consecutivas
filtro3
```

```python
filtro3.to_csv("filtro3.csv")
from google.colab import files
files.download("filtro3.csv")
```

```python
#SE APLICA FILTRO QUE EL INDICE CVE_DOC ENCUENTRE 2 F37048 Y F37046 Y NOS DE LA COLUMNA PRECIO_UNITARIO
df1=pd.read_excel("Detalle precios y productos fabricados 2022.xlsx",index_col=0)
df1
filtro4=df1.loc[["F37048","F37046"], ["PRECIO_UNITARIO"]]
filtro4
```

```python
filtro4.to_csv("filtro4.csv")
from google.colab import files
files.download("filtro4.csv")
```

```python
#SE APLICA FILTRO PARA MOSTRAR LAS PRIMERAS 5 LINEAS
filtro5=df.head()
filtro5
```

```python
filtro5.to_csv("filtro5.csv")
from google.colab import files
files.download("filtro5.csv")
```

```python
#SE APLICA FILTRO PARA MOSTRAR LAS ULTIMAS 5 LINEAS

filtro6=df.tail()
filtro6
```

```python
filtro6.to_csv("filtro6.csv")
from google.colab import files
files.download("filtro6.csv")
```

```python
#SE APLICA FILTRO EN LA COLUMNA PRECIO UNITARIO PARA MOSTRAR LAS QUE SEAN MAS DE 700
filtro7=df[df["PRECIO_UNITARIO"] > 700]
filtro7
```

```python
filtro7.to_csv("filtro7.csv")
from google.colab import files
files.download("filtro7.csv")
```

```python
#SE APLICA FILTRO EN LA COLUMNA PRECIO UNITARIO PARA MOSTRAR LAS QUE SEAN MAS DE 700 Y EN
#LA COLUMNA NOMBRE_VENDEDOR PARA LAS QUE SEAN IGUALES A LETICIA RAMIREZ HERNANDEZ
filtro8=df[(df["COSTO_UNITARIO"] > 700) & (df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
filtro8
```

```python
filtro8.to_csv("filtro8.csv")
from google.colab import files
files.download("filtro8.csv")
```

```python
#SE APLICA FILTRO EN LA COLUMNA PRECIO UNITARIO PARA MOSTRAR LAS QUE SEAN MAS DE 700 O EN
#LA COLUMNA NOMBRE_VENDEDOR PARA LAS QUE SEAN IGUALES A LETICIA RAMIREZ HERNANDEZ
filtro9=df[(df["COSTO_UNITARIO"] > 700)  | (df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
filtro9
```

```python
filtro9.to_csv("filtro9.csv")
from google.colab import files
files.download("filtro9.csv")
```

```python
# SE APLICA UN FILTRO PARA QUE SEA LO CONTRARIO DEL PASADO
filtro10=df[~(df["COSTO_UNITARIO"] >= 700)  & ~(df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
filtro10
```

```python
filtro10.to_csv("filtro10.csv")
from google.colab import files
files.download("filtro10.csv")
```
